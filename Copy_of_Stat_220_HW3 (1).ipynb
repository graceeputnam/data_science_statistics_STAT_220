## Problem 1
Let X be a normal random variable with mean 8 and standard deviation 4. Find the following probabilities:

```python
mean = 8
std = 4
```

A. $Pr( X > 10)$

```python
pr_a = 1 - norm.cdf(10, mean, std)
print(pr_a)
```

B. Pr(X > 10 and X < 12)

```python
pr_b = norm.cdf(12, mean, std) - norm.cdf(10, mean, std)
print(pr_b)
```

C. Pr(X > 10|X < 12)

```python
pr_c = pr_b/ norm.cdf(12, mean, std)
print(pr_c)
```

D. Pr(X > 10|X > 8)

```python
pr_d = pr_a / (1- norm.cdf(8, mean, std))
print(pr_d)
```

E. Pr(X > 10 and X > 8)

## Problem 2
